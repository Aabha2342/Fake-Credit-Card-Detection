Importing the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading the dataset to a Pandas DataFrame


```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

Dataset Information

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

Distribution of Legit and Fraudulent Transactions

```python
credit_card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0-->Normal Transactions

1-->Fraudelent Transactions

```python
#seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the value for both transcations
credit_card_data.groupby('Class').mean()
```

UNDER-SAMPLING

build a sample dataset containing similar distribution of normal and fraudulent transactions

Number of fraud transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating the two samples

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the dataset into Features

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting in training testing data
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify =Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Training Data",training_data_accuracy)
```

```python
#accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Test Data",testing_data_accuracy)
```
